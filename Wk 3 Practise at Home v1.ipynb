```python
import csv

filepath = r'C:\Users\techw_000\Desktop\foodhygienedata.csv'

rat1=0
rat2=0
rat3=0
rat4=0
rat5=0
rat0=0

with open (filepath,"r") as file:
    reader = csv.DictReader(file,delimiter = ",")
    for row in reader:
        if row["rating"] == "1":
            rat1 = rat1 + 1
        if row["rating"] == "2":
            rat2 = rat2 + 1
        if row["rating"] == "3":
            rat3 = rat3 + 1
        if row["rating"] == "4":
            rat4 = rat4 + 1
        if row["rating"] == "5":
            rat5 = rat5 + 1
        if row["rating"] == "":
            rat0 = rat0 + 1
            

print("Amount Scoring 1:",rat1)
print("Amount Scoring 2:",rat2)
print("Amount Scoring 3:",rat3)
print("Amount Scoring 4:",rat4)
print("Amount Scoring 5:",rat5)
print("Amount missing:",rat0)
```

```python
import csv
import datetime as dt

today = dt.date.today() 
#inspdate =[]
rec =[]
fileread = r'C:\Users\techw_000\Desktop\foodhygienedata.csv'
filewrite = r'C:\Users\techw_000\Desktop\foodhygienedata_Inspectedsince.csv'

header=["EstablishmentName","InspectionDate","DaysSinceInspection"]
with open (filewrite,"w",newline="") as writehead:
    recval = csv.writer(writehead,delimiter=",")
    recval.writerow(header)
    
with open(fileread,"r") as file:
    reader = csv.DictReader(file,delimiter=",")
    for row in reader:
        if row["inspectiondate"] == "":
            #print("record skipped no inspection date")
            delta="record skipped no inspection date"
        else:
            #print(row["inspectiondate"])
            dd,mm,yyyy=row["inspectiondate"].split("/")
            delta = today - dt.date(int(yyyy),int(mm),int(dd))
            #print(delta)
        rec = [row["establishmentname"],row["inspectiondate"],delta]
        with open (filewrite,"a",newline="") as writedet:
            recval = csv.writer(writedet,delimiter=",")
            recval.writerow(rec)
```
