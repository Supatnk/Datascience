```python
#A count of how many establishments received each rating: 1, 2, 3, 4 and 5, 
#and how many ratings were omitted. Output should look similar to the following
import csv
r1=0
r2=0
r3=0
r4=0
r5=0
miss=0

filepath = r'C:\Users\B00749292\Desktop\foodhygienedata.csv'

with open(filepath,'r') as file:
    readcsv = csv.DictReader(file,delimiter = ',')  
    for row in readcsv:
        if row["rating"] == '1':
            r1 += 1
        if row["rating"] == '2':
            r2 += 1
        if row["rating"] == '3':
            r3 += 1
        if row["rating"] == '4':
            r4 += 1
        if row["rating"] == '5':
            r5 += 1
        if row["rating"] == '':
            miss += 1
            
print("Amount Scoring:",r1)
print("Amount Scoring:",r2)
print("Amount Scoring:",r3)
print("Amount Scoring:",r4)
print("Amount Scoring:",r5)
print("Amount missing:",miss)
```

```python
#Read in the establishment names and inspection dates listed in the Food Hygiene Dataset
#For each record, calculate the value “DaysSinceInspection”. 
#This value contains the number of days since the date specified in the “inspectiondate” field.
#Create a new CSV file which contains a collection of establishment names, inspection dates, and days since inspection. 

from datetime import date
import csv
import datetime

filepath = r'C:\Users\B00749292\Desktop\foodhygienedata.csv'

#headers =[h1,h2,h3]
values = [0,0,0]
d0=[0,0,0]


with open(filepath,'r') as file:
    readcsv = csv.DictReader(file,delimiter = ',')
    writer = csv.writer(file,delimiter=',')
    #writer = csv.writer(headers)
    for row in readcsv:
        values[0] = row["establishmentname"]
        values[1] = row["inspectiondate"]
        d0 = row["inspectiondate"].split("/")
        dt=date(int(d0[2]),int(d0[1]),int(d0[0]))
        delta = date.today() - dt
        print(values[0],values[1],delta.days)
```
