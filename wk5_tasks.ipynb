```python
from bokeh.plotting import output_notebook,show,figure
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.sampledata.autompg2 import autompg2 as mpg
from bokeh.layouts import gridplot

output_notebook()

data = ColumnDataSource(data=mpg)
data_group = ColumnDataSource(data=mpg.groupby("displ"))

tools="box_select"

p = figure(height=300,width=450,title='Engine Capacity vs City MPG',x_axis_label='Engine capacity',y_axis_label='City MPG',tools=tools)
p2 = figure(height=300,width=450,x_range=p.x_range,y_range=p.y_range,title='Engine Capacity vs Highway MPG',x_axis_label='Engine capacity',y_axis_label='Highway MPG')

p.square(x='displ',y='cty',source=data)
p2.circle(x='displ',y='hwy',source=data)

p.line(x='displ',y='cty_mean',source=data_group,color="red")
p2.line(x='displ',y='hwy_mean',source=data_group,color="red")

gplot=gridplot([[p,p2]])
show(gplot)
```
